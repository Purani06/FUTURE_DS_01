```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/Sample - Superstore.csv", encoding="latin1")

category_profit = data.groupby('Category')['Profit'].sum()
category_profit_k = category_profit / 1000

ax = category_profit_k.plot(kind='bar')

plt.title('Profit by Category (in K)')
plt.xlabel('Category')
plt.ylabel('Profit (K)')

plt.show()
```

```python
region_profit = data.groupby('Region')['Profit'].sum()
region_profit
```

```python
region_profit = data.groupby('Region')['Profit'].sum()
```

```python
region_profit_k = region_profit / 1000
region_profit_k.plot(kind='bar')

import matplotlib.pyplot as plt

plt.title('Profit by Region (in K)')
plt.xlabel('Region')
plt.ylabel('Profit (K)')
plt.show()
```
